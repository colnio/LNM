```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
data = pd.read_csv("2.csv")
data 
```

```python
x = [float(i[:-1:]) for i in data.keys()[1::]]
x.sort()
x = np.array(x)
y = np.array(data['Vsd'])
```

```python
data = data.drop(columns=['Vsd'])
```

```python
data = data.reindex(sorted(data.columns, key=lambda x : float(x[:-1:])), axis=1)
```

```python
extent = [x[0], x[-1], y[0], y[-1]]
```

```python
plt.figure(figsize=(20, 10))

plt.clf()
plt.imshow(data.values.T, extent=extent, origin='lower', interpolation='bilinear')
plt.colorbar()
# plt.savefig("f.png", dpi=400)
plt.show()
```
