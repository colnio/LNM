```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as si
import pandas as pd 
```

```python
data = pd.read_csv("2.csv")
data 
```

```python
x = [float(i[:-1:]) for i in data.keys()[1::]]
x.sort()
x = np.array(x)
y = np.array(data['Vsd']) * 1000
```

```python
data = data.drop(columns=['Vsd'])
```

```python
data = data.reindex(sorted(data.columns, key=lambda x : float(x[:-1:])), axis=1)
```

```python
extent = [x[0], x[-1], y[0], y[-1]]
```

```python
plt.figure(figsize=(15, 15))

plt.clf()
plt.imshow(data.values, extent=extent, origin='lower', interpolation='bilinear', aspect='auto')
cb = plt.colorbar()
cb.set_label('dI/dV', rotation=270)
plt.title("dI/dV vs (Vsd, B)")
plt.rcParams.update({'font.size': 25})
plt.xlabel('Field, T')
plt.ylabel('Vsd, mV')
# plt.savefig("f.png", dpi=400)
plt.show()
```
